**<font size=4 color = #FF7F00> - 問題：</font>**

windows檔案路徑長相：  
    "E:\III\III\workplace\topic\pyWebcrawler\fatsecret\fatsecret_food\fatsecret_foodData\fatsecret_foodData.json"

```python
path = "..\zTestdata\testdata_foodData.json"
df = pd.read_json(path, encoding="utf-8")
df.head()
```

**<font size=4 color = #FF7F00> - 原因：</font>**

    路徑中的”\”為python 的跳脫字元，故導致python編譯錯誤。

**<font size=4 color = #FF7F00> - 解決方法：</font>**

    將"”\”修改為"\\"或"/"即可。

#### 解法一：

在路徑加 r，其作用是將後面的字做為正規表達式（regular expression），在此可想成類似跳脫字元（\），避免路徑有"\"而被python當作跳脫字元。

```python
path = r"..\zTestdata\testdata_foodData.json"
df = pd.read_json(path, encoding="utf-8")
df.head()
```

#### 解法二：

利用"\"跳脫字元之作用，將後面的"\"作為正常的"\"路徑表示法。

```python
path = "..\\zTestdata\\testdata_foodData.json"
df = pd.read_json(path, encoding="utf-8")
df.head()
```

#### 解法三：

利用類似linux之路徑表示法。

```python
path = "../zTestdata/testdata_foodData.json"
df = pd.read_json(path, encoding="utf-8")
df.head()
```
