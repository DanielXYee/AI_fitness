```python
import os
import csv
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
files= os.listdir(r"./img_dataset/train")
        
for a in files:
    img = cv2.imread("./img_dataset/train"+"/"+a, 0)
    plt.imshow(img, cmap ='gray')
    plt.show()


    tag = int(input("0:thin,1:medium,2:large"))

    with open('trainlabels.csv', 'a', newline='') as csvfile:
      # 建立 CSV 檔寫入器
      writer = csv.writer(csvfile)
      # 寫入一列資料
      writer.writerow([tag,])

    continue
```
